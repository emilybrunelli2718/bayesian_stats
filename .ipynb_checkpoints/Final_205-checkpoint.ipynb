**1.** I will be using a Bayesian multiple regression model with hormone therapy, age, smoking, drinking, statins, diabetes, and BMI considered as parameters in order to predict low-density lipoprotein cholesterol levels after 1 year of enrollment in the HERS study. I will use DIC as my model selection criteria. This will be useful for making inferences about subpopulations, as well as examining the effect of any of the covariates on LDL. 

```r
library("ggplot2")
library("readxl")    # alternative installation of the %>%
library(magrittr) # needs to be run every time you start R and want to use %>%
library(dplyr)    # alternatively, this also loads %>%
library(R2jags)
library(ggplot2)
library(cowplot)
library(jtools)
```

```r
data<-read.csv("hersdata_1.csv")
data[data==""]<-NA
data = data[complete.cases(data), ]
data = data[!(data$LDL1 <= 0),]
```

```r
data <- data %>%
      mutate(smoking = ifelse(smoking == "no",0,1))

data <- data %>%
      mutate(drinkany = ifelse(drinkany == "no",0,1))

data <- data %>%
      mutate(exercise = ifelse(exercise == "no",0,1))

data <- data %>%
      mutate(statins = ifelse(statins == "no",0,1))

data <- data %>%
      mutate(diabetes = ifelse(diabetes == "no",0,1))
```

```r
data <- data %>%
      mutate(HT = ifelse(HT == "placebo",0,1))
```

**3.** How are the g-prior and independent prior different?

The g-prior falls under the "empirical Bayes" category because your prior is constructed from the data you have, where the independent prior has nothing to do with the data. Another way they are different is that the g-prior is an improper prior, the independent prior is a proper prior. This can be problematic because it can complicate model selection. The g-prior (unsurprisingly) has results that tend to depend a lot on your specific value of g, from what I understand. 

**4.** Discuss possible models for the HERS data, including different sets of predictors and prior construction. 

I will be using an independent prior for simplicity, $beta[i] \sim dnorm(0, .001)$, and using DIC to compare them. I will only be looking at main effects. Hypothetically one could iterate through all the possible models, because there are only $2^8 - 1$ that aren't trivial, but I'll just be checking them by hand-selecting models to try (partially because I'm not sure how to do this easily in R). 

As a baseline/for reference it's good to look at the full model, so I started with that and removed variables I suspected weren't very helpful (like drinking and exercise), based on the results of the exploratory data analysis. I also trimmed down the model to only HT + statins to verify that a more complex model was in fact nessecary. (The minimalist model does almost as poorly as the full model). 

Based on DIC values, the best model uses HT, BMI, smoking, statins, and diabetes as predictors. Interestingly, it does not include age (one might expect that LDL gets worse as you get older, for example) but when considering that the slope of the line for LDL $\sim$ age is close to flat, perhaps this should not be surprising. Drinking and exercise are excluded from the best models.

**FULL MODEL**

```r
jags.data=list(n = nrow(data), LDL1 = data$LDL1, HT=data$HT, age=data$age, smoking=data$smoking, drinkany = data$drinkany, exercise = data$exercise, statins = data$statins, diabetes = data$diabetes, BMI = data$BMI)
model= "model {
for(i in 1:n) {
    LDL1[i] ~ dnorm(mu[i],tau)
    mu[i] <- beta[1] + beta[2]*HT[i] + beta[3]*age[i] + beta[4]*smoking[i] +  beta[5]*drinkany[i] + beta[6]*exercise[i] + beta[7]*statins[i] + beta[8]*diabetes[i] + beta[9]*BMI[i]
}
for(i in 1:9){
    beta[i] ~ dnorm(0,0.001)
}
tau ~ dgamma(0.001,0.001)

}"
jags.inits <- list(list(tau=1,beta=c(0,0,0,0, 0, 0, 0, 0 ,0)))
jags.param <- c("beta","tau")
model.fit <- jags(jags.data, jags.inits, jags.param,model.file=textConnection(model),n.chains=1, n.iter=12000, n.thin=1, n.burnin=2000)
jags.mcmc=as.mcmc(model.fit)
plot(jags.mcmc)
summary(jags.mcmc)
full <- model.fit$BUGSoutput$DIC
print(full)
```
